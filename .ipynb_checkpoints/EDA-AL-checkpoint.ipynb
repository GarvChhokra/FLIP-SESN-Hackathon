```python
# This notebook performs the following:
# Data analysis on the Canada's medication and reaction dataset
# Clustering of the dataset using aggleromative clustering
import pandas as pd


# load the data
df = pd.read_csv('output.csv', header=0)
print('First 15 records:')
print(df.head(15))
print(f'DF shape: {df.shape}')

```

```python
df.info()
```

```python
# make a df where age is not null
df_age_not_null = df[df['age'].notnull()]
# drop the age outliers
df_age_not_null = df_age_not_null[df_age_not_null['age'] < 110]
# check for min max of age
print(f'Age min: {df_age_not_null["age"].min()}')
print(f'Age max: {df_age_not_null["age"].max()}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Grouping the data by 'report_id'
df_grouped = df_age_not_null.groupby('report_id').agg({
    'age': 'mean',
    'weight': 'mean',
    'gender_eng': 'first',
    'pt_name_eng': lambda x: x.mode()[0] if not x.mode().empty else x.iloc[0],
    'drugname': lambda x: ', '.join(set(x))  # Combine all drug names associated with a report_id
}).reset_index()

# Plot: Distribution of Age by Reaction (Grouped by Report ID)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_grouped, x='pt_name_eng', y='age')
plt.xticks(rotation=90)
plt.xlabel('Fall-related Reaction')
plt.ylabel('Age')
plt.title('Distribution of Age by Fall-related Reaction (Grouped by Report ID)')
plt.tight_layout()
plt.show()

# Plot: Top Drugs Prescribed
top_drugs = df['drugname'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_drugs.values, y=top_drugs.index)
plt.xlabel('Number of Reports')
plt.ylabel('Drug Name')
plt.title('Top 10 Drugs Prescribed')
plt.tight_layout()
plt.show()

# Plot: Gender Distribution for Adverse Reactions
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='pt_name_eng', hue='gender_eng', order=df['pt_name_eng'].value_counts().nlargest(10).index)
plt.xticks(rotation=90)
plt.xlabel('Fall-related Reaction')
plt.ylabel('Number of Reports')
plt.title('Top 10 Fall-related Reactions by Gender')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

# Plot: Scatter Plot of Weight vs. Age (Grouped by Report ID)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_grouped, x='age', y='weight', hue='gender_eng')
plt.xlabel('Age')
plt.ylabel('Weight (kg)')
plt.title('Scatter Plot of Weight vs. Age (Grouped by Report ID)')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# plot to visualize the top 10 active ingredients related to adverse reactions
top_drugs = df['active_ingredient_name'].value_counts().nlargest(15)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_drugs.values, y=top_drugs.index)
plt.xlabel('Number of Reports')
plt.ylabel('Active Ingredient')
plt.title('Top 15 Active Ingredients Related to Adverse Reactions')
plt.tight_layout()
plt.show()


```

```python
# preprocessing
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline


# Drop unique ID columns
columns_to_drop = ['report_id', 'report_no', 'reaction_id', 'active_ingredient_id']
df = df_age_not_null.drop(columns=columns_to_drop)

# Label encoding for columns with a large number of unique values
label_encoder_active_ing = LabelEncoder()
label_encoder_drugname = LabelEncoder()

# Fit and transform the labels
df['active_ingredient_name'] = label_encoder_active_ing.fit_transform(df['active_ingredient_name'])
df['drugname'] = label_encoder_drugname.fit_transform(df['drugname'])

# Update the categorical features list (excluding the label encoded columns)
categorical_features = ['weight_unit_eng', 'gender_eng', 'pt_name_eng', 'soc_name_eng']
numerical_features = ['age', 'weight']

# Define preprocessing for numerical and categorical features with missing value handling
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Impute missing values with mean for numerical features
    ('scaler', StandardScaler())
])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Impute missing values with the most frequent value for categorical features
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)

# Sample the dataset to reduce its size
df_sampled = df.sample(frac=0.01, random_state=42)  # Use 10% of the data
# Preprocess the sampled data
df_preprocessed = preprocessor.fit_transform(df_sampled)

# Convert sparse matrix to dense for clustering
df_preprocessed_dense = df_preprocessed if isinstance(df_preprocessed, np.ndarray) else df_preprocessed.toarray()

# Clustering Algorithms to Experiment With
clustering_algorithms = {
    'AgglomerativeClustering': AgglomerativeClustering(n_clusters=5),
    'DBSCAN': DBSCAN(eps=0.5, min_samples=5)
}

# Experiment with Different Clustering Algorithms
for name, algorithm in clustering_algorithms.items():
    # Fit the clustering algorithm
    labels = algorithm.fit_predict(df_preprocessed_dense)
    
    # Calculate and print silhouette score if applicable
    if len(set(labels)) > 1 and -1 not in set(labels):  # Silhouette score is not defined for single cluster or all noise
        silhouette_avg = silhouette_score(df_preprocessed_dense, labels)
        print(f"Silhouette Score for {name}: {silhouette_avg}")
        # show the number of clusters
        print(f"Number of clusters for {name}: {len(set(labels))}")
    else:
        print(f"Silhouette Score for {name}: Not applicable (single cluster or all noise)")

# Add the cluster labels from AgglomerativeClustering to the original sampled DataFrame
df_sampled['cluster'] = clustering_algorithms['AgglomerativeClustering'].fit_predict(df_preprocessed_dense)

#
# Save the resulting DataFrame with cluster labels to a new CSV file
output_path = 'clustered_output_sampled.csv'
df_sampled.to_csv(output_path, index=False)

# Display the first few rows of the sampled DataFrame with the cluster labels
print(df_sampled.head())
```

```python
# visualize the clusters from agglomerative clustering
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_sampled, x='age', y='weight', hue='cluster', palette='tab10')
plt.xlabel('Age')
plt.ylabel('Weight (kg)')
plt.title('Clusters from Agglomerative Clustering')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```
